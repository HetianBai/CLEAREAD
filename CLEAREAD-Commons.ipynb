```python
filename = input('please enter the name of the book for the commonroom')
```

```python
word_list = []
word_list_clean = []
File = open( filename + ".txt", "r",encoding = 'utf-8')
for i in File:
    word_list.append(i.rstrip())
for i in word_list:
    if i != '':
        word_list_clean.append(i)
```

```python
word_cn = []
word_pron_char=[]
```

```python
app_id = '4e001e20'
app_key = '4647a64fc8110e4304392175bf4782cd'
for i in word_list_clean:
    try:
        word_id = re.findall("[a-zA-Z]",i)[0]
        strictMatch = 'false'

        url = 'https://od-api.oxforddictionaries.com:443/api/v2/entries/en-gb/' + word_id.lower() + '?strictMatch=' + strictMatch;

        r = requests.get(url, headers = {'app_id': app_id, 'app_key': app_key})
        pron = r.json()["results"][0]['lexicalEntries'][0]['entries'][0]['pronunciations'][0]['phoneticSpelling']
        char = r.json()["results"][0]['lexicalEntries'][0]['lexicalCategory']['text']
        word_pron_char.append([i,'/'+pron+'/',char])
        time.sleep(5)
        print(i)
    except KeyError:
        pass
    continue

for i in word_pron_char:
    if i[2] == 'Noun':
        i[2] = 'n.'
    if i[2] == 'Adjective':
        i[2] = 'adj.'
    if i[2] == 'Verb':
        i[2] = 'v.'
    if i[2] == 'Adverb':
        i[2] = 'adv.'
```
